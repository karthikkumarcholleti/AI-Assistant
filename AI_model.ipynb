**Model 1 - distilbert-base-uncased-distilled-squad**

```python
from transformers import pipeline
# Load a pre-trained model
pipeline = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")

# Example context
context = """
Apple Inc. is trading at $150 as of today. The opening price was $145, the highest price duringthe day was $152,
and the lowest was $144. The company saw significant growth in its tech portfolio.
"""
question = "What is the current price of Apple?"
# answer
answer = pipeline(question=question, context=context)
print(f"Answer: {answer['answer']}")
```

**Model 3- T5 (Text-to-text Transfer Transormer) Model**

```python
from transformers import pipeline

# Load a T5 model for text generation
t5_pipeline = pipeline("text2text-generation", model="t5-small")

# Define the context
context = """
Apple Inc. Stock Information:
- Current price: $150
- Opening price: $145
- Highest price: $150
- Lowest price: $140
"""

input_text = f"Answer the question based on the following information: {context} Question: What was the highest price of Apple today?"
#response
response = t5_pipeline(input_text)
print(response[0]['generated_text'])
```

```python
from transformers import pipeline

t5_pipeline = pipeline("text2text-generation", model="t5-small")

context = """
Apple Inc. Stock Information:
- Current price: $150
- Opening price: $145
- Highest price: $150
- Lowest price: $140
"""
input_text = f"Answer the question based on the following information: {context} Question: What was the highest price of Apple today?"

response = t5_pipeline(input_text, max_new_tokens=50)

# Print the generated response
print(response[0]['generated_text'])
```
